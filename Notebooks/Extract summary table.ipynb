```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
file = "../Exported Results/Results/Monthly_results/stats/aggregated_stats.txt"
```

```python
def get_value_key(line_split):
    key, value = line_split.split(":")
    return key.replace("▸", ""), value.rstrip().strip()
```

```python
f = open(file, "r")
instr_dict = {}
for line in f:
    if "=========" in line:
        instrument = [s.strip() for s in line.split("============")][1]
        instr_dict[instrument] = {}

    elif line.startswith('▸'):
        if len(line.split(' - '))>1:
            for values in line.split(' - '):
                key, value = get_value_key(values)
                instr_dict[instrument][key] = value

        else:
            if '▸Power Law K-test' in line.split(' - ')[0]:
                continue

            key, value = get_value_key(line.split(' - ')[0])           
            instr_dict[instrument][key] = value
            
for key in instr_dict:
    instr_dict[key]['Is Weakly connected?'], instr_dict[key]['Is Strongly connected?'] = instr_dict[key]['Is Strongly connected?'], instr_dict[key]['Is Weakly connected?']
```

```python
stats = pd.DataFrame.from_dict(instr_dict, orient='index')
```
